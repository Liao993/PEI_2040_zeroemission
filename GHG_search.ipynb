```python
import pandas as pd
```

```python
test3 = pd.read_csv("data/GHGIPCC.csv", encoding='latin-1')
```

```python
PEI_GHG = test3[test3['Region'] == 'Prince Edward Island']
```

```python
PEI_GHG.Category.unique()
```

```python
re = PEI_GHG[(PEI_GHG['Category'] == 'Residential') & (PEI_GHG['Year'] >= 1999)]
```

```python
PEI_GHG.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
year = re['Year'].astype('int64')
co2 = re['CO2eq'].astype('float64')
```

```python
plt.figure(figsize=(26,4))
ax = sns.lineplot(x=year, y=co2)

plt.title("The CO2 equivalent from residential places in PEI from 1999 to 2019", fontsize=30)

plt.xlabel("Year", fontsize=25)
plt.xticks(rotation=20, fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel("CO2eq",fontsize=25)

plt.savefig('GHG_fromhouse.png')
```

## percentage

```python
total = PEI_GHG[(PEI_GHG['Category'] == 'TOTAL') & (PEI_GHG['Year'] >= 1999)]
```

```python
co2t = total['CO2eq'].astype('float64')
```

```python
percentage = [a / b for a, b in zip(list(co2), list(co2t))]
```

```python
plt.figure(figsize=(26,4))
ax2 = sns.lineplot(x=year, y=percentage)

plt.title("The percentage of GHG emission from residential places in PEI from 1999 to 2019", fontsize=30)

plt.xlabel("Year", fontsize=25)
plt.xticks(rotation=20, fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel("CO2eq",fontsize=25)

plt.savefig('GHG_percenthouse.png')
```
